  - ìµœê·¼ ë°ì´í„°ì™€ ê³¼ê±° ë°ì´í„°ë¥¼ ë™ì¼í•œ ê°€ì¤‘ì¹˜ë¡œ ì·¨ê¸‰í•˜ë¯€ë¡œ ìµœì‹  ë°ì´í„°ì˜ ì¤‘ìš”í•œ íŠ¹ì„±ì„ ë°˜ì˜í•˜ì§€ ëª» í•  ìˆ˜ ìžˆìŒ

$MA_t = \frac{1}{n} \sum_{i=0}^{n-1} Y_{t-i} = \frac{Y_t + Y_{t-1} + \dots + Y_{t-n+1}}{n}$

- $MA_t$: $t$ ì‹œì ì˜ ì´ë™ í‰ê·  ê°’
- $Y_t$: $t$ ì‹œì ì˜ ì‹¤ì œ ê°’
- $n$: ìœˆë„ìš° í¬ê¸° (í‰ê· ì„ ê³„ì‚°í•˜ëŠ” ê¸°ê°„)

#### ì§€ìˆ˜ ê°€ì¤‘ ì´ë™ í‰ê·  (Exponential Weighted Moving Average, EWMA) 
- ê³¼ê±° ë°ì´í„°ì— ì§€ìˆ˜ì ìœ¼ë¡œ ê°ì†Œí•˜ëŠ” ê°€ì¤‘ì¹˜ë¥¼ ë¶€ì—¬í•˜ëŠ” ê¸°ë²• (ìµœì‹  ë°ì´í„°ê°€ í‰ê· ê°’ì— ë” í° ì˜í–¥ì„ ë¯¸ì¹¨)
  - ìµœê·¼ ë°ì´í„°ì— ë” ë¯¼ê°í•˜ê²Œ ë°˜ì‘í•˜ë©° ë‹¨ìˆœ ì´ë™ í‰ê· ë³´ë‹¤ ë°ì´í„°ì˜ ë³€í™”ë¥¼ ë” ìž˜ ë°˜ì˜í•¨
- Exponential smoothing(ì§€ìˆ˜ í‰í™œ)ì´ë¼ê³ ë„ í•¨

$EWM_t = \alpha Y_t + (1-\alpha) EWM_{t-1}$

- $EWM_t$: $t$ ì‹œì ì˜ ì´ë™ í‰ê·  ê°’
- $Y_t$: $t$ ì‹œì ì˜ ì‹¤ì œ ê°’
- $\alpha$: í‰í™œ ê³„ìˆ˜ ($ 0 <$ $\alpha$ $< 1$)

#### ë‹¨ìˆœ ì´ë™ í‰ê·  ê³„ì‚° ì˜ˆì œ 1
- DataFrameì˜ rolling() í•¨ìˆ˜ ì´ìš©
  - ì´ë™ ìœˆë„ìš° ê¸°ë°˜ì˜ í†µê³„ ê³„ì‚° ì§€ì›
  - ë°˜í™˜ê°’ì„ ì´ìš©í•´ ì§‘ê³„ í•¨ìˆ˜(mean(), sum(), min(), max() ë“±)ë¥¼ í˜¸ì¶œí•˜ì—¬ ì›í•˜ëŠ” ê²°ê³¼ ê³„ì‚°
  - ì‹œê³„ì—´ ë°ì´í„° ë¶„ì„ì˜ í•µì‹¬ ë„êµ¬
  - https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.rolling.html

```python
# ìƒ˜í”Œ ë°ì´í„° ìƒì„±
data = {'Value': [10, 20, 30, 40, 50, 60, 70]}
df = pd.DataFrame(data)

# 3ê°œ ê°’ì˜ ì´ë™ í‰ê·  ê³„ì‚°
df['Moving_Avg'] = df['Value'].rolling(window=3).mean() # rolling() í•¨ìˆ˜ì˜ window ë§¤ê°œë³€ìˆ˜ë¡œ ìœˆë„ìš° í¬ê¸°ë¥¼ ì§€ì •
print(df)

print()

# 3ê°œ ê°’ì˜ ì´ë™ í•©ê³„ ê³„ì‚°
df['Moving_Sum'] = df['Value'].rolling(window=3).sum()
print(df)
```

#### ë‹¨ìˆœ ì´ë™ í‰ê·  ê³„ì‚° ì˜ˆì œ 2

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# ì‹œê³„ì—´ ë°ì´í„° ìƒì„±
date_rng = pd.date_range(start='2023-01-01', end='2023-03-31', freq='D') # 2023ë…„ 1ì›” 1ì¼ë¶€í„° 2023ë…„ 3ì›” 31ì¼ê¹Œì§€ ì¼('D') ë‹¨ìœ„ ë‚ ì§œ ìƒì„±
data = np.random.randint(50, 150, len(date_rng)) # ìœ„ ë‚ ì§œ ê¸¸ì´ë§Œí¼ì˜ 50ë¶€í„° 150 ì‚¬ì´ì˜ ëžœë¤ ì •ìˆ˜ê°’ ìƒì„±
df = pd.DataFrame({'Date': date_rng, 'Value': data}) # ìœ„ ë‚ ì§œì™€ ë°ì´í„°ë¡œ DataFrame ê°ì²´ ìƒì„±
df.set_index('Date', inplace=True) # Date ì»¬ëŸ¼ì„ ì¸ë±ìŠ¤ë¡œ ì„¤ì •

# ìœˆë„ìš° í¬ê¸° 7ì¸ ë‹¨ìˆœ ì´ë™ í‰ê·  ê³„ì‚°
df['SMA_7'] = df['Value'].rolling(window=7).mean()

# ì‹œê°í™”
plt.figure(figsize=(10, 4))
plt.plot(df['Value'], label='Original Data')
plt.plot(df['SMA_7'], label='7-Day Moving Average', color='red')
plt.legend()
plt.show()
```

#### ì§€ìˆ˜ ê°€ì¤‘ ì´ë™ í‰ê·  ê³„ì‚° ì˜ˆì œ 1
- DataFrameì˜ ewm() í•¨ìˆ˜ ì´ìš©
- rolling() í•¨ìˆ˜ì™€ ìœ ì‚¬í•˜ê²Œ ë°˜í™˜ê°’ì— í†µê³„ í•¨ìˆ˜(mean(), sum(), var() ë“±)ë¥¼ í˜¸ì¶œí•˜ì—¬ ê²°ê³¼ë¥¼ ê³„ì‚°
- https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.ewm.html

```python
DataFrame.ewm(com=None, span=None, halflife=None, alpha=None, min_periods=1, adjust=True, ignore_na=False, axis=0)
```

##### í‰í™œê³„ìˆ˜ ì„¤ì • ë°©ë²•
- **`alpha ì§ì ‘ ì„¤ì •`**: $ \alpha = \text{ì‚¬ìš©ìžê°€ ì§€ì •í•œ ê°’} $
- **`com`**: $ \alpha = \frac{1}{1 + \text{com}} $
- **`span`**: $ \alpha = \frac{2}{\text{span} + 1} $
- **`halflife`**: $ \alpha = 1 - e^{-\ln(2) / \text{halflife}} $

```python
import pandas as pd

# ìƒ˜í”Œ ë°ì´í„° ìƒì„±
data = [10, 20, 30, 40, 50, 60, 70]
df = pd.DataFrame({'Value': data})

# EWMA ê³„ì‚° (span=3)
df['EWMA'] = df['Value'].ewm(span=3, adjust=False).mean() # spanì´ 3ì´ë©´ aëŠ” 0.5
print(df)
# EWMA ì´ˆê¸°ê°’ì€ ì› ê°’ ê·¸ëŒ€ë¡œ ìœ ì§€
```
