- 간단한 그래프 그리기 예제

##### 선 그래프 예제 1
- 그래프 제목과 x, y축 레이블 추가
- 한글 폰트 사용

- 제목과 x축, y축 레이블 설정
  - title(), xlabel(), ylabel() 함수

- 한글 폰트 사용
  - rc 모듈의 rc() 함수

```python
import matplotlib.pyplot as plt

from matplotlib import rc

rc('font', family='Malgun Gothic') # font로 맑은 고딕 폰트를 설정 - 여기서는 영문으로 사용해야 함

# 기온 변화 데이터
time = ['6시', '9시', '12시', '15시', '18시', '21시']
temperature = [15, 18, 22, 25, 21, 18]

# 선 그래프 그리기
plt.plot(time, temperature)

# 그래프 제목 및 레이블 설정
plt.title('시간에 따른 기온 변화', fontsize=16)
plt.xlabel('시간', fontsize=12)
plt.ylabel('기온(℃)', fontsize=12)
plt.grid(True) # 그래프 내에 격자선 표시

# 그래프 보여주기
plt.show()
```

##### 선 그래프 예제 2
- 범례 (legend) 표시
  - legend() 함수

```python
import matplotlib.pyplot as plt
from matplotlib import rc

rc('font', family='Malgun Gothic')

# 기온 변화 데이터
time = ['6시', '9시', '12시', '15시', '18시', '21시']
temperature_summer = [15, 18, 22, 25, 21, 18]
temperature_winter = [1, 3, 6, 7, 4, 2]

# 선 그래프 그리기
plt.plot(time, temperature_summer, label='Summer')
plt.plot(time, temperature_winter, label='Winter')

# 그래프 제목 및 레이블 설정
plt.title('시간에 따른 기온 변화', fontsize=16)
plt.xlabel('시간', fontsize=12)
plt.ylabel('기온(℃)', fontsize=12)
plt.grid(True) # 그래프 내에 격자선 표시

# 범례(legend)를 화면에 표시, loc 파라미터 - 범례의 위치 설정
plt.legend(loc='upper right') 
# https://matplotlib.org/3.1.1/api/_as_gen/matplotlib.pyplot.legend.html

# 그래프 보여주기
plt.show()
```

##### 선 그래프 예제 3
- 선 스타일 설정
  - plot() 함수의 color, linestyle, marker 파라미터

```python
import matplotlib.pyplot as plt
from matplotlib import rc

rc('font', family='Malgun Gothic')

# 기온 변화 데이터
time = ['6시', '9시', '12시', '15시', '18시', '21시']
temperature_summer = [15, 18, 22, 25, 21, 18]
temperature_winter = [1, 3, 6, 7, 4, 2]
temperature_fall = [3, 7, 9, 10, 6, 4]
temperature_spring = [8, 10, 13, 15, 11, 7]

# 선 그래프 그리기
plt.plot(time, temperature_spring, label='Spring', color='#00ffff', linestyle='--', marker='x') # '-.' 일점쇄선, 's' square
plt.plot(time, temperature_summer, label='Summer', color='red', linestyle='-', marker='o') # '-' 실선, 'o' circle
plt.plot(time, temperature_fall, label='Fall', color='#00ff00', linestyle=':', marker='d') # ':' 점선, 'd' diamond
plt.plot(time, temperature_winter, label='Winter', color='magenta', linestyle='-.', marker='s') # '--' 파선, 'x' x

# 그래프 제목 및 레이블 설정
plt.title('시간에 따른 기온 변화', fontsize=16)
plt.xlabel('시간', fontsize=12)
plt.ylabel('기온(℃)', fontsize=12)
plt.grid(True) # 그래프 내에 격자선 표시

# 범례(legend)를 화면에 표시, loc 파라미터 - 범례의 위치 설정
plt.legend(loc='upper left') 
# https://matplotlib.org/3.1.1/api/_as_gen/matplotlib.pyplot.legend.html

# 그래프 보여주기
plt.show()
```

##### 선 그래프 예제 4
- x, y 축 범위와 눈금 설정
  - xlim(), ylim(), xticks(), yticks() 함수
  - xticks(ticks=None, labels=None, **kwargs) https://matplotlib.org/3.3.2/api/_as_gen/matplotlib.pyplot.xticks.html
    - ticks: x축 눈금 위치를 나타내는 리스트 / labels: x축 눈금 레이블 (ticks 파라미터가 있는 경우만 가능)
  - grid() 함수

```python
import matplotlib.pyplot as plt

xdata = list(range(30))
ydata = []

for x in xdata:
    y = 2 * x
    ydata.append(y)

plt.plot(xdata, ydata, label='y=2x')
plt.legend(loc='best')
plt.xlim(0, 35) # x축의 범위
plt.ylim(0, 70) # y축의 범위
plt.xticks(list(range(0, 35, 2))) # x축의 눈금
plt.yticks(list(range(0, 70, 5))) # y축의 눈금

plt.grid(True) # 그래프에 그리드 표시
#plt.grid(False) # 그리드 표시하지 않음
plt.show()
```

#### 파이 그래프 (Pie chart)
- 전체 중에서 각 항목이 차지하는 비율을 시각적으로 보여줄 때 적합함
- 전체를 원으로 표현하고 원 안의 각 조각이 항목의 비율을 나타내므로 부분과 전체 간의 관계를 한눈에 파악할 수 있음

##### 파이 그래프를 활용하는 경우
- 각 항목이 전체에서 차지하는 비율을 보여주는 경우
  - 시장 점유율, 예산의 항목별 분배, 설문 조사 응답 비율
- 전체 데이터가 몇 개의 주요 부분으로 나뉠 때 그 구성을 표현하는 경우
  - 회사의 총 매출에서 각 제품이 차지하는 비중, 시간 분배 (하루 일과에서 업무, 수면, 여가 등)
- 데이터 항목이 적을 때 (5~7개 이하) 효과적임
  - 항목이 너무 많으면 그래프가 복잡해져 비율을 시각적으로 구분하기 어려워짐

##### 파이 그래프 (Pie chart) 예제
- pie() 함수
  - https://matplotlib.org/3.3.2/api/_as_gen/matplotlib.pyplot.pie.html#matplotlib.pyplot.pie

```python
import matplotlib.pyplot as plt
from matplotlib import rc

rc('font', family='Malgun Gothic')

pets = ['개', '고양이', '기타', '기르지않음']
portion = [18, 3.4, 3.1, 75.5 ]

plt.title('총 가구 대비 반려동물 기르는 비율(2018)')

plt.pie(portion, explode=(0, 0.1, 0, 0), labels=pets, autopct='%1.1f%%', shadow=True, startangle=90)
#plt.pie(portion, explode=(0, 0, 0, 0), labels=pets, autopct='%1.1f%%', shadow=True, startangle=90) # explode 
#plt.pie(portion, explode=(0, 0, 0, 0), labels=pets, autopct='%1d%%', shadow=True, startangle=90) # autopct 숫자 표시 형식
#plt.pie(portion, explode=(0, 0, 0, 0), labels=pets, autopct='%1d%%', shadow=False, startangle=90) # shadow 그림자
#plt.pie(portion, explode=(0, 0, 0, 0), labels=pets, autopct='%1d%%', shadow=True, startangle=0) # startangle 시작 각도, 반시계 방향으로 

plt.show()
```

#### 히스토그램 (Histogram)
- 데이터의 분포를 시각화하는 데 적합함
- 특히, 연속형 데이터의 값을 구간(bins)으로 나누어 그 값들이 얼마나 자주 발생했는지 보여줌
- 데이터의 분포 형태, 중심 경향, 분산도, 이상치를 파악하는 데 유용함
- 단, 데이터의 구간 설정에 따라 결과가 달라질 수 있어 구간 선택이 분석의 중요한 요소이고 개별 데이터 포인트에 대한 구체적인 정보는 제공하지 않고 데이터의 전반적인 분포만 보여줄 수 있음

##### 히스토그램을 활용하는 경우
- 연속형 데이터의 값을 여러 구간으로 나누어 빈도수를 확인하는 경우
  - 시험 점수, 나이, 키, 몸무게, 온도
- 데이터의 분포(특정 구간에 집중되어 있는지, 고르게 퍼져 있는지 등)를 분석하는 경우
  - 시험 점수, 상품 구매 가격, 주택 가격
- 데이터의 중심 및 변동성을 파악하는 경우
- 이상치를 식별하는 경우
  - 데이터가 일반적인 분포에서 크게 벗어난 값을 쉽게 찾을 수 있음

##### 히스토그램 예제 1
- hist() 함수
  - https://matplotlib.org/3.3.3/api/_as_gen/matplotlib.pyplot.hist.html

```python
import random
import matplotlib.pyplot as plt

lst = [random.randint(1,100) for i in range(300)]
plt.hist(lst, bins=10) # bins - 히스토그램을 나누는 구간 수

plt.show()
```

##### 히스토그램 예제 2

```python
import numpy as np
import matplotlib.pyplot as plt

from matplotlib import rc

rc('font', family='Malgun Gothic')

# 상품 가격 데이터 (랜덤으로 100개의 데이터 생성, 10 ~ 60 사이)
np.random.seed(0)  # 재현성을 위해 시드 설정
prices = np.random.randint(10, 61, 100) # 10에서 60 사이의 상품 가격을 임의로 100개 생성

# 그래프 제목 및 레이블 설정
plt.title('상품 가격 분포', fontsize=16)
plt.xlabel('가격 (만원)', fontsize=12)
plt.ylabel('빈도 (상품 수)', fontsize=12)
plt.grid(axis='y')

# 히스토그램 그리기
plt.hist(prices, bins=10, color='skyblue', edgecolor='black') # 그래프의 색상은 skyblue, 막대 테두리 색상은 black

# 그래프 보여주기
plt.show()
```

##### 상자 그림 (box plot) 예제 2

```python
import matplotlib.pyplot as plt
from matplotlib import rc

rc('font', family='Malgun Gothic')

# 시험 점수 데이터 (예시 데이터)
scores = [
    [88, 70, 85, 82, 75, 91, 87, 84, 89, 93],  # 1반
    [68, 82, 75, 80, 77, 85, 79, 83, 71, 76],  # 2반
    [65, 70, 68, 82, 74, 69, 71, 87, 66, 73],  # 3반
    [75, 88, 82, 91, 84, 92, 63, 87, 90, 81],  # 4반
]

# 박스 플롯 그리기
plt.boxplot(scores, tick_labels=['1반', '2반', '3반', '4반'])

# 그래프 제목 및 레이블 설정
plt.title('반별 시험 점수 분포', fontsize=16)
plt.xlabel('반', fontsize=12)
plt.ylabel('시험 점수', fontsize=12)
plt.grid(True)

# 그래프 보여주기
plt.show()
```

##### 그래프 이미지 파일로 저장
- savefig() 함수

```python
import matplotlib.pyplot as plt
from matplotlib import rc

rc('font', family='Malgun Gothic')

x = ['영희', '철수', '재호']
y = [7, 3, 5]
plt.title('연간 영화관람 횟수')
plt.xlabel('이름')
plt.ylabel('횟수')

plt.bar(x, y) # 막대(바) 그래프

plt.savefig('fig1.png', dpi=300) 
# 저장 가능 포맷 - png, pdf, svg
# dpi 파라미터는 생략 가능

plt.show()
```
